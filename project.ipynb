```python
import pandas as pd
import numpy as np
import sklearn 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Dry_Bean_Dataset.csv')
print(data.info())
```

```python
data.hist(bins = 40, figsize = (20,15))
```

```python

our_data, validator_data = train_test_split(data, test_size = 0.2, random_state = 42)
our_data.to_csv('our_data.csv', index = False)
validator_data.to_csv('validator_data.csv', index = False)
```

```python
df = our_data
```

```python
df.hist(bins = 40, figsize = (20,15))
```

```python
#rozkad shapefactor wzgldem klasy
plt.figure(figsize=(10, 6))
sns.violinplot(x='Class', y='ShapeFactor1', data=df)
```
